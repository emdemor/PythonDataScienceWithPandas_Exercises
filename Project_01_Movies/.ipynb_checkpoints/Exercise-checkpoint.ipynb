# Data Import and First Inspection

Importing the dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Setting Pandas

```python
# Shot up to 30 columns
pd.options.display.max_columns = 30

# Float with just two decimal places
pd.options.display.float_format = '{:,.2f}'.format
```

Reading the dataset

```python
# importing the dataset, forcing tha pandas
# to treate the data on column 'release_date'
# as date format
df = pd.read_csv('movies_complete.csv',
                 parse_dates = ['release_date'])

df.columns = ['', 'Title', 'Tagline', 'Date', 'Genres',
       'Collection', 'Language', 'Budget',
       'Revenue', 'Companies', 'Countries',
       'Votes', 'Average Rating', 'Popularity', 'Runtime', 'Overview',
       'Spoken Languages', 'Poster', 'Cast', 'Cast Size', 'Crew Size',
       'Director']
```

```python
# create a sample for fast tests
df_sample = df.sample(n=100,random_state=1)
```

# Most Produced Genre

```python
nTotal = len(df)
df_genres = df['Genres'].str.split('|',expand=True).stack().reset_index(level=1,drop=True).to_frame()
df_genres = df_genres[0].value_counts().to_frame()
df_genres[0] = 100.*df_genres[0]/nTotal
df_genres.columns = ['Percentual']
plt1 = df_genres.plot.bar(y='Percentual',
                          rot=90,
                          title='Percentual of movies with the genre tag',
                          legend=False)

plt1.set_ylabel("%")
plt1.get_figure().savefig('genre_percent.png', bbox_inches='tight')
```

```python
df_genres = df['Genres'].str.split('|',expand=True).stack().reset_index(level=1,drop=True).to_frame()
genre_list = df_genres.drop_duplicates(subset=0, keep="last").reset_index()[0].tolist()

df_genre = df.copy()

conditions  = True
conditions &= df_genre['Genres'].notna()
conditions &= df_genre['Date'].notna()
df_genre = df_genre[conditions]


df_genre['Year'] = pd.DatetimeIndex(df_genre['Date']).year
df_genre['Month'] = pd.DatetimeIndex(df_genre['Date']).month

df_new = df_genre[['Year']]

for genre in genre_list:
    
    df_new.insert(len(df_new.columns),
                    genre,
                    df_genre['Genres'].str.contains(genre).astype(int).tolist(),
                    allow_duplicates = True) 
    
```

```python
plt2 = df_new.groupby('Year')[genre_list].sum().plot(y=genre_list,
                                              figsize=(10,7),
                                              title='Genre of movies over years')
plt2.set_ylabel("Amount of movies");

plt2.get_figure().savefig('genre_amount_over_years.png', bbox_inches='tight')
```

# Budget and Revenue over Years

```python
conditions  = True
conditions &= df['Date'].notna()
conditions &= df['Budget'].notna()

df_date = df[conditions].copy()

df_date['Year'] = pd.DatetimeIndex(df_date['Date']).year
df_date['Month'] = pd.DatetimeIndex(df_date['Date']).month

df_year = df_date['Month'].value_counts().to_frame().sort_values('Month')
df_year.columns = ['Number of Movies']
df_year = df_year.sort_index()
#df_year.plot.bar( y='Number of Movies', rot=70, figsize=(20,12))
```

```python
plt3 = df_date.groupby('Year')['Budget'].mean().to_frame().plot(y='Budget',
                                                                title = 'Budget over years',
                                                                legend = False)
plt3.set_ylabel("Mean Budget (millions of USD)");
plt3.get_figure().savefig('budget.png', bbox_inches='tight')
```

```python
test = pd.merge(
    df_date.groupby('Year')['Budget'].mean(),
    df_date.groupby('Year')['Revenue'].mean(),
    how = 'left', on = 'Year'
)
test.columns = ['Budget','Revenue']

plt4 = test.plot(y=test.columns,
                 title = 'Revenue and Budget over years')
plt4.set_ylabel("Mllions of USD");
plt4.get_figure().savefig('budget_revenue.png', bbox_inches='tight')
```
