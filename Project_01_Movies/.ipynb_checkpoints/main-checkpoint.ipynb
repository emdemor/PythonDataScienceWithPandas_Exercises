# The Best and the Worst Movie

```python
from IPython.display import HTML
```

```python
df_best = df[['poster_path',
           'title',
           'budget_musd',
           'revenue_musd',
           'vote_count',
           'vote_average',
           'popularity',
           'release_date']].copy()


# profits is the difference between revenue and budget
#df['profit_musd'] = df.revenue_musd.sub(df.budget_musd)
df_best['profit_musd'] = df_best['revenue_musd']-df_best['budget_musd']

# return precentual
df_best['return'] = df_best['revenue_musd']/df_best['budget_musd']

# Changing the names of columns
df_best.columns = ['',
                   'Title',
                   'Budget',
                   'Revenue',
                   'Votes',
                   'Average Rating',
                   'Popularity',
                   'Date',
                   'Profit',
                   'ROI' ]

# Changind the index
df_best.set_index('Title',inplace = True)

# Filling non numerical values with 0
#for column in ['Budget','Average Rating','Revenue']:
#    df_best[column].fillna(0,inplace = True)
```

```python
df_best.info()
```

```python
def checking_intervals(df_best,intervals = []):
    
    range_list = []
    
    if len(intervals) > 0:
        f_type = type(min)
        df_inter = pd.DataFrame(intervals)
        for col in df_best.columns.tolist()[1:]:

            clmn = df_inter.loc[df_inter[0] == col]

            min_list = clmn[1].tolist()
            max_list = clmn[2].tolist()

            minValue = None
            maxValue = None

            while min in min_list: min_list.remove(min)
            if len(min_list) == 0:
                minValue = min
                min_list = [min]

            while max in min_list: min_list.remove(max)
            if len(min_list) == 0: minValue = max

            while max in max_list: max_list.remove(max)
            if len(max_list) == 0:
                maxValue = max
                max_list = [max]

            while min in max_list: max_list.remove(min)
            if len(max_list) == 0: maxValue = min

            if minValue == None: minValue = max(min_list)
            if maxValue == None: maxValue = min(max_list)

            if(minValue == max or maxValue == min):
                range_list.append([col,maxValue,minValue])

            else:
                range_list.append([col,minValue,maxValue])

        
    return range_list


def filtering_movies(df_best,conditional_array):
    df_best_cpy = df_best.copy()
    df_Interval = df_best_cpy.describe().loc[['min','max']]
        
    try: 
        for cond in conditional_array: 
            column,minValue,maxValue = cond
            if not (minValue == min):  df_Interval[column]['min'] = minValue
            if not (maxValue == max): df_Interval[column]['max'] = maxValue
    except:
        print('>>> Error on conditions.')

    df_best_cpy = df_best.copy()
    conditions  = True
    for column in df_Interval.columns.tolist():
        conditions  &= df_best_cpy[column] >= df_Interval[column]['min']
        conditions  &= df_best_cpy[column] <= df_Interval[column]['max']
        
    return df_best_cpy[conditions]


def classification(df_best,by,intervals=[],quantiles=[],ascending=False):
    '''
    Its possible to use this function with the original dataframe to
    
    '''
    listQuant = []
    
    for elem in quantiles:
        col,minQ,maxQ = elem
        non_null = df_best[df_best[col] != 0][col].dropna()
        if( (0 <= minQ <= 1.) & (0 <= maxQ <= 1.) ):
            listQuant.append([col,non_null.quantile(minQ),non_null.quantile(maxQ)])
            
        
    return filtering_movies(df_best,
                            checking_intervals(df_best,intervals = intervals+listQuant)).sort_values(by=by,ascending=ascending)

def best_worst(df_best,by,n=5,intervals=[],ascending=False,quantiles=[]):
    dataframe = classification(df_best,by=by,intervals=intervals,quantiles=quantiles,ascending=ascending)
    subset = dataframe[['',by]]
    return HTML(subset.iloc[:n].to_html(escape=False))
    
    
```
